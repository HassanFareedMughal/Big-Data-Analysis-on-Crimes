```python
unique_years = df['Year'].unique()

for year in unique_years:
    year_data = df[df['Year'] == year]

    murders = year_data[year_data['Offence'] == 'Murder']

    regions = ['Punjab', 'Sindh', 'KP', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']
    murder_cases = murders[regions].iloc[0]

    plt.figure(figsize=(10, 6))
    plt.bar(regions, murder_cases)
    plt.title(f'Number of Murders in Different Regions of Pakistan ({year})')
    plt.xlabel('Regions')
    plt.ylabel('Number of Murders')

    for i, v in enumerate(murder_cases):
        plt.text(i, v, str(v), ha='center', va='bottom')

    plt.show()
```
