```python
ax = sns.countplot(x = 'Punjab', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
violence = df.groupby(['Punjab'], as_index = False)['Year'].sum().sort_values(by = 'Year', ascending = False)
sns.barplot(x = 'Punjab', y = 'Year', data = violence)
```

```python
data = pd.DataFrame({
    'Region': ['Punjab', 'Sindh', 'Balochistan'],
    'Metric': ['Murders', 'Murders', 'Murders'],
    'Value': [100, 150, 50]
})

murder_data = data[data['Metric'] == 'Murders']

colors = ['blue', 'green', 'red']

plt.bar(murder_data['Region'], murder_data['Value'], color=colors)

plt.xlabel('Region')
plt.ylabel('Number of Murders')
plt.title('Murders in Different Regions')
plt.show()

```
